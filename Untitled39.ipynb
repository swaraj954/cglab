```python
#midpoint circle
import matplotlib.pyplot as plt
import math

def draw_circle(xc, yc, r):
    x = 0
    y = r
    d = 1 - r

    plot_points(xc, yc, x, y)

    while x < y:
        x += 1
        if d < 0:
            d += 2 * x + 1
        else:
            y -= 1
            d += 2 * (x - y) + 1
        plot_points(xc, yc, x, y)

def plot_points(xc, yc, x, y):
    plt.plot(xc+x, yc+y, 'go')
    plt.plot(xc-x, yc+y, 'go')
    plt.plot(xc+x, yc-y, 'go')
    plt.plot(xc-x, yc-y, 'go')
    plt.plot(xc+y, yc+x, 'go')
    plt.plot(xc-y, yc+x, 'go')
    plt.plot(xc+y, yc-x, 'go')
    plt.plot(xc-y, yc-x, 'go')


x=int(input("Enter x :"))
y=int(input("Enter y :"))
r=int(input("Enter radius r :"))
draw_circle(x,y,r)
plt.axis('scaled')
```

```python
#bresenham circle
import matplotlib.pyplot as plt

def plot_point(xc,yc,x,y):
    plt.plot(xc+x,yc+y,'go')
    plt.plot(xc-x,yc+y,'go')
    plt.plot(xc+x,yc-y,'go')
    plt.plot(xc-x,yc-y,'go')
    plt.plot(xc+y,yc+x,'go')
    plt.plot(xc-y,yc+x,'go')
    plt.plot(xc+y,yc-x,'go')
    plt.plot(xc-y,yc-x,'go')
    
def bresenham_circle(xc,yc,r):
    x=0
    y=r
    d=3-2*r
    plot_point(xc,yc,x,y)
    while(x<y):
        x+=1
        if(d<0):
           d+=4*x+6
        else:
            y-=1
            d+=4*(x-y)+10
        plot_point(xc,yc,x,y)

x=int(input())
y=int(input())
r=int(input())
bresenham_circle(x,y,r)
plt.xlim([-100,100])
plt.ylim([0,100])
#plt.tight_layout()
plt.axis('scaled')
```

```python
#point clipping
import matplotlib.pyplot as plt
fig,ax=plt.subplots()

def pointClip(XY, n, Xmin, Ymin, Xmax, Ymax):
 
    ax.axhline(y=Xmin,color='r')
    ax.axhline(y=Xmax,color='r')
    ax.axvline(x=Ymin,color='r')
    ax.axvline(x=Ymax,color='r')
    print("Point inside the viewing pane:")
    for i in range(n):
        if ((XY[i][0] >= Xmin) and
            (XY[i][0] <= Xmax)):
            if ((XY[i][1] >= Ymin) and
                (XY[i][1] <= Ymax)):
                plt.scatter(XY[i][0],XY[i][1],c='blue')
                
         
    print("\n\nPoint outside the viewing pane:")
    for i in range(n):    
        if ((XY[i][0] < Xmin) or (XY[i][0] > Xmax)) :
            plt.scatter(XY[i][0],XY[i][1],c='orange')
        if ((XY[i][1] < Ymin) or (XY[i][1] > Ymax)) :
            plt.scatter(XY[i][0],XY[i][1],c='orange')
    plt.show()
 
if __name__ == '__main__':
    Xmin = int(input("Xmin="))
    Xmax = int(input("Xmax="))
    Ymin = int(input("Ymin="))
    Ymax = int(input("Ymax="))
    XY=[]
    n=int(input("Number of Points-->"))
    for i in range(n):
        x=int(input(f"X{i}->"))
        y=int(input(f"Y{i}->"))
        XY.append([x,y])
    pointClip(XY, len(XY), Xmin, Ymin, Xmax, Ymax)
```

```python
#line cohensuther
import matplotlib.pyplot as plt

INSIDE = 0  
LEFT = 1  
RIGHT = 2  
BOTTOM = 4  
TOP = 8  

def compute(x, y):
    code = INSIDE
    if x < xmin:
        code |= LEFT
    elif x > xmax:
        code |= RIGHT
    if y < ymin:
        code |= BOTTOM
    elif y > ymax:
        code |= TOP

    return code


def cohenSutherlandClip(x1, y1, x2, y2):
    p = compute(x1, y1)
    q = compute(x2, y2)
    accept = False

    while True:

        if p == 0 and q == 0:
            accept = True
            break

        elif (p & q) != 0:
            break

        else:
            x = 1.0
            y = 1.0
            if p != 0:
                out = p
            else:
                out = q

            if out & TOP:
                x = x1 + (x2 - x1) * (ymax - y1) / (y2 - y1)
                y = ymax

            elif out & BOTTOM:
                x = x1 + (x2 - x1) * (ymin - y1) / (y2 - y1)
                y = ymin

            elif out & RIGHT:
                y = y1 + (y2 - y1) * (xmax - x1) / (x2 - x1)
                x = xmax

            elif out & LEFT:
                y = y1 + (y2 - y1) * (xmin - x1) / (x2 - x1)
                x = xmin

            if out == p:
                x1 = x
                y1 = y
                p = compute(x1, y1)

            else:
                x2 = x
                y2 = y
                q = compute(x2, y2)

    if accept:
        print("\n\nLine accepted from %.2f, %.2f to %.2f, %.2f" % (x1, y1, x2, y2))
        plt.scatter(x1, y1, color="red")
        plt.scatter(x2, y2, color="red")
        a=[x1,x2]
        b=[y1,y2]
        plt.plot(a, b, color="green")
        plt.text(x1,y1,'({}, {})'.format(int(x1), int(y1)))
        plt.text(x2,y2,'({}, {})'.format(int(x2), int(y2)))
    else:
        print("Line not present in the window")

   
if __name__ == '__main__':
    print("Xmin : ")
    xmin = int(int(input()))
    print("Xmax : ")
    xmax = int(int(input()))
    print("Ymin : ")
    ymin = int(int(input()))
    print("Ymax : ")
    ymax = int(int(input()))

    x = [xmin, xmax, xmax, xmin, xmin]
    y = [ymin, ymin, ymax, ymax, ymin]

    print("\nEnter x1 : ")
    x1 = int(int(input()))
    print("Enter y1 : ")
    y1 = int(int(input()))
    print("Enter x2 : ")
    x2= int(int(input()))
    print("Enter y2 : ")
    y2 = int(int(input()))

    cohenSutherlandClip(x1, y1, x2, y2)

    plt.plot(x, y)
    plt.axis('scaled')
    plt.show()
```

```python
#line liang
import matplotlib.pyplot as plt

def liang(x1,y1,x2,y2,xmin,ymin,xmax,ymax):
    dx=x2-x1
    dy=y2-y1
    
    p1=-dx
    p2=dx
    p3=-dy
    p4=dy
    
    q1=x1-xmin
    q2=xmax-x1
    q3=y1-ymin
    q4=ymax-y1
    
    t1=q1/p1
    t2=q2/p2
    t3=q3/p3
    t4=q4/p4
    
    tmax=0
    tmin=1
    
    if(p1==0 or p2==0 or p3==0 or p4==0):
        print('line parallel')
    if(p1!=0):
        if(p1<0):
            tmax=max(tmax,t1)
        elif(p1>0):
            tmin=min(tmin,t1)
    if(p2!=0):
        if(p2<0):
            tmax=max(tmax,t2)
        elif(p2>0):
            tmin=min(tmin,t2)
    if(p3!=0):
        if(p3<0):
            tmax=max(tmax,t3)
        elif(p3>0):
            tmin=min(tmin,t3)
    if(p4!=0):
        if(p4<0):
            tmax=max(tmax,t4)
        elif(p4>0):
            tmin=min(tmin,t4)    
            
    if(tmax>tmin):
        print("line completely outside")
        
    if(tmax<tmin):
        x_1=x1+(tmax*dx)
        y_1=y1+(tmax*dy)
        x_2=x1+(tmin*dx)
        y_2=y1+(tmin*dy)
    print('line will be accepted from (%.2f,%.2f)to(%.2f,%.2f)'%(x_1,y_1,x_2,y_2))    
    plt.scatter(x_1,y_1,color='red')
    plt.scatter(x_2,y_2,color='red')
    plt.text(x_1,y_1,'({},{})'.format(int(x_1),int(y_1)))
    plt.text(x_2,y_2,'({},{})'.format(int(x_2),int(y_2)))
    
xmin=int(input())    
ymin=int(input())    
xmax=int(input())    
ymax=int(input())    

x=[xmin,xmax,xmax,xmin,xmin]
y=[ymin,ymin,ymax,ymax,ymin]

x1=int(input())    
y1=int(input())    
x2=int(input())    
y2=int(input()) 

liang(x1,y1,x2,y2,xmin,ymin,xmax,ymax)
a=[x1,x2]
b=[y1,y2]
plt.plot(x,y)
plt.plot(a,b,color='green')
plt.axis('scaled')
plt.show()
```

```python
#polygon clipping
import matplotlib.pyplot as plt
def sutherland_hodgman_clip(subject_polygon, clip_polygon):
    def inside(p):
        return (cp2[0]-cp1[0])*(p[1]-cp1[1]) > (cp2[1]-cp1[1])*(p[0]-cp1[0])

    def compute_intersection():
        dc = [cp1[0] - cp2[0], cp1[1] - cp2[1]]
        dp = [s[0] - e[0], s[1] - e[1]]
        n1 = cp1[0] * cp2[1] - cp1[1] * cp2[0]
        n2 = s[0] * e[1] - s[1] * e[0]
        n3 = 1.0 / (dc[0] * dp[1] - dc[1] * dp[0])
        return [(n1*dp[0] - n2*dc[0]) * n3, (n1*dp[1] - n2*dc[1]) * n3]

    output_list = subject_polygon
    cp1 = clip_polygon[-1]

    for cp2 in clip_polygon:
        input_list = output_list
        output_list = []
        s = input_list[-1]

        for e in input_list:
            if inside(e):
                if not inside(s):
                    output_list.append(compute_intersection())
                output_list.append(e)
            elif inside(s):
                output_list.append(compute_intersection())
            s = e
        cp1 = cp2
    return output_list

subject_polygon = [(100, 120), (220, 250), (330,320), (380, 400), (420, 450),(100,120)]
clip_polygon = [(150, 150), (350, 150), (350, 350), (150, 350)]

clipped_polygon = sutherland_hodgman_clip(subject_polygon, clip_polygon)
print(clipped_polygon)
x, y = zip(*clipped_polygon)

fig, ax = plt.subplots()

ax.plot(*zip(*subject_polygon), color='blue', label='Input polygon')
ax.plot(x, y, color='red', label='Clipped polygon')

x_min = 150
y_min = 150
x_max = 350
y_max = 350

plt.plot([x_min, x_min], [y_min, y_max], 'go', linestyle = '--') #left
plt.plot([x_min, x_max], [y_min, y_min], 'go', linestyle = '--') #bottom
plt.plot([x_min, x_max], [y_max, y_max], 'go', linestyle = '--') #top
plt.plot([x_max, x_max], [y_max, y_min], 'go', linestyle = '--') #right

ax.legend()
ax.set_title('Sutherland-Hodgman clipping algorithm')

plt.show()
```
